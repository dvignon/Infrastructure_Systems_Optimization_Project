```python
import networkx as nx
import matplotlib.pyplot as plt  

import pandas as pd
import numpy as np
from scipy import sparse
```

```python
g=nx.Graph()
```

```python
param = pd.read_csv("Data/Anaheim/Anaheimflow.csv", header = 0)
param.head()
```

```python
g.add_nodes_from([1,416])
```

```python
#od=[""]
for i in param.index:
    #od.extend([(param.Tail[i],param.Head[i])])
    g.add_edge(param.Tail[i],param.Head[i],weight=param.Volume)
```

```python
nx.draw_circular(g)
```

```python
plt.show() # display  
```
